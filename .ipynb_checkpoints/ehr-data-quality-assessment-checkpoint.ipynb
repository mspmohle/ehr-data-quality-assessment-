```python
# ==========================================================
# 2. Synthetic EHR Data Generation
# ==========================================================
# Purpose:
#   - Create a synthetic electronic health record (EHR) dataset
#   - Include typical clinical fields: demographics, encounters, vitals, labs
#   - Intentionally introduce small data quality issues for testing
# ==========================================================

import pandas as pd
import numpy as np
import os

# --- Directory setup ---------------------------------------------------------
os.makedirs("../data/raw", exist_ok=True)
data_path = "../data/raw/synthetic_ehr_data.csv"

# --- Random seed for reproducibility -----------------------------------------
np.random.seed(42)

# --- Define record count -----------------------------------------------------
n = 1000

# --- Generate synthetic EHR data ---------------------------------------------
ehr_df = pd.DataFrame({
    "patient_id": [f"P{str(i).zfill(5)}" for i in range(1, n + 1)],
    "age": np.random.randint(0, 100, n),
    "sex": np.random.choice(["M", "F", None], n, p=[0.48, 0.48, 0.04]),
    "encounter_date": pd.date_range("2023-01-01", periods=n, freq="D"),
    "encounter_type": np.random.choice(
        ["Inpatient", "Outpatient", "Emergency"], n, p=[0.3, 0.6, 0.1]
    ),
    "diagnosis_code": np.random.choice(
        ["E11.9", "I10", "J45.909", "M54.5", None], n, p=[0.25, 0.25, 0.25, 0.2, 0.05]
    ),
    "height_cm": np.random.normal(170, 10, n).round(1),
    "weight_kg": np.random.normal(75, 15, n).round(1),
    "systolic_bp": np.random.normal(120, 15, n).round(0),
    "diastolic_bp": np.random.normal(80, 10, n).round(0),
    "cholesterol_mg_dl": np.random.normal(190, 35, n).round(0)
})

# --- Introduce data quality issues -------------------------------------------
# Randomly inject missingness and impossible values
ehr_df.loc[np.random.choice(ehr_df.index, 10), "age"] = None
ehr_df.loc[np.random.choice(ehr_df.index, 5), "weight_kg"] = -5  # impossible weights
ehr_df.loc[np.random.choice(ehr_df.index, 5), "systolic_bp"] = 400  # outliers

# --- Save synthetic dataset --------------------------------------------------
ehr_df.to_csv(data_path, index=False)
print(f"✅ Synthetic EHR dataset saved → {data_path}")

# --- Quick preview -----------------------------------------------------------
print("\nPreview:")
display(ehr_df.head())
```

```python
# ==========================================================
# 3. Data Quality Profiling
# ==========================================================
# Purpose:
#   - Assess completeness, validity, and consistency of EHR dataset
#   - Use Great Expectations 0.18+ API for validation
#   - Identify missing, invalid, or extreme values
# ==========================================================

import pandas as pd
import great_expectations as ge
from great_expectations.dataset import PandasDataset
import numpy as np
import os

# --- Load dataset ------------------------------------------------------------
df = pd.read_csv("../data/raw/synthetic_ehr_data.csv")
ge_df = ge.dataset.PandasDataset(df)  # backwards-compatible wrapper

print(f"✅ Loaded dataset → {df.shape[0]} rows, {df.shape[1]} columns\n")

# --- Basic data completeness metrics -----------------------------------------
missing_summary = df.isna().sum().sort_values(ascending=False)
missing_percent = (missing_summary / len(df) * 100).round(1)

print("Missing Values Summary (Top 10):")
display(pd.DataFrame({
    "Missing Count": missing_summary,
    "Missing %": missing_percent
}).head(10))

# --- Numeric range validation (Great Expectations) ---------------------------
results = {
    "age_valid": ge_df.expect_column_values_to_be_between("age", 0, 110),
    "weight_valid": ge_df.expect_column_values_to_be_between("weight_kg", 0, 300),
    "systolic_valid": ge_df.expect_column_values_to_be_between("systolic_bp", 60, 200),
    "diastolic_valid": ge_df.expect_column_values_to_be_between("diastolic_bp", 40, 120),
}

print("\nValidation Results:")
for test, result in results.items():
    print(f" - {test}: {'✅ Passed' if result.success else '❌ Failed'} "
          f"({result.result['unexpected_count']} unexpected values)")

# --- Outlier detection summary -----------------------------------------------
numeric_cols = ["height_cm", "weight_kg", "systolic_bp", "diastolic_bp", "cholesterol_mg_dl"]
summary = df[numeric_cols].describe().T
summary["IQR"] = summary["75%"] - summary["25%"]
summary["Lower_Bound"] = summary["25%"] - 1.5 * summary["IQR"]
summary["Upper_Bound"] = summary["75%"] + 1.5 * summary["IQR"]

print("\nOutlier Thresholds (IQR-based):")
display(summary[["mean", "std", "Lower_Bound", "Upper_Bound"]])

# --- Save summary artifacts --------------------------------------------------
os.makedirs("../data/processed", exist_ok=True)
summary_path = "../data/processed/data_quality_summary.csv"
summary.to_csv(summary_path)
print(f"\n✅ Data quality summary saved → {summary_path}")
```

```python
# ==========================================================
# 3. Data Quality Profiling (Pandas-based)
# ==========================================================
# Purpose:
#   - Assess completeness, validity, and consistency of the EHR dataset
#   - Identify missing, invalid, or extreme values using pandas logic
#   - Output results to processed directory for later visualization
# ==========================================================

import pandas as pd
import numpy as np
import os

# --- Load dataset ------------------------------------------------------------
df = pd.read_csv("../data/raw/synthetic_ehr_data.csv")
print(f"✅ Loaded dataset → {df.shape[0]} rows, {df.shape[1]} columns\n")

# --- Basic data completeness metrics -----------------------------------------
missing_summary = df.isna().sum().sort_values(ascending=False)
missing_percent = (missing_summary / len(df) * 100).round(1)
missing_df = pd.DataFrame({
    "Missing Count": missing_summary,
    "Missing %": missing_percent
})
print("Missing Values Summary (Top 10):")
display(missing_df.head(10))

# --- Numeric range validation ------------------------------------------------
validation_rules = {
    "age": (0, 110),
    "weight_kg": (0, 300),
    "systolic_bp": (60, 200),
    "diastolic_bp": (40, 120)
}

print("\nValidation Results:")
validation_results = []
for col, (low, high) in validation_rules.items():
    if col in df.columns:
        invalid_mask = (df[col] < low) | (df[col] > high)
        invalid_count = invalid_mask.sum()
        validation_results.append((col, invalid_count))
        print(f" - {col}: {'✅ Passed' if invalid_count == 0 else '❌ Failed'} ({invalid_count} out of range)")
    else:
        print(f" - {col}: ⚠️ Column not found in dataset")

validation_df = pd.DataFrame(validation_results, columns=["Column", "Out_of_Range_Count"])

# --- Outlier detection summary -----------------------------------------------
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
summary = df[numeric_cols].describe().T
summary["IQR"] = summary["75%"] - summary["25%"]
summary["Lower_Bound"] = summary["25%"] - 1.5 * summary["IQR"]
summary["Upper_Bound"] = summary["75%"] + 1.5 * summary["IQR"]

print("\nOutlier Thresholds (IQR-based):")
display(summary[["mean", "std", "Lower_Bound", "Upper_Bound"]].head(10))

# --- Save summary artifacts --------------------------------------------------
os.makedirs("../data/processed", exist_ok=True)
summary_path = "../data/processed/data_quality_summary.csv"
validation_path = "../data/processed/data_validation_summary.csv"
missing_path = "../data/processed/missing_values_summary.csv"

summary.to_csv(summary_path)
validation_df.to_csv(validation_path, index=False)
missing_df.to_csv(missing_path)

print(f"\n✅ Data quality summaries saved to ../data/processed/")
```

```python
# ==========================================================
# 4. Data Visualization and Anomaly Highlighting (Display + Save)
# ==========================================================
# Purpose:
#   - Visualize numeric distributions and missing-value patterns
#   - Highlight anomalies or extreme values
#   - Show plots inline and save to ../figures/
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --- Load data ---------------------------------------------------------------
df = pd.read_csv("../data/raw/synthetic_ehr_data.csv")
os.makedirs("../figures", exist_ok=True)

# --- Style setup -------------------------------------------------------------
plt.style.use("ggplot")
sns.set_palette("muted")

# --- 1. Distribution plots for key numeric variables -------------------------
numeric_cols = ["age", "weight_kg", "height_cm", "systolic_bp", "diastolic_bp", "cholesterol_mg_dl"]

for col in numeric_cols:
    if col in df.columns:
        plt.figure(figsize=(8, 5))
        sns.histplot(df[col], bins=30, kde=True)
        plt.title(f"Distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.tight_layout()

        # Display inline
        plt.show()

        # Save figure
        fname = f"../figures/distribution_{col}.png"
        plt.savefig(fname)
        plt.close()
        print(f"✅ Saved {fname}")

# --- 2. Missing-value heatmap -----------------------------------------------
plt.figure(figsize=(10, 6))
sns.heatmap(df[numeric_cols].isna(), cbar=False, cmap="viridis")
plt.title("Missing Value Heatmap (Key Numeric Features)")
plt.xlabel("Columns")
plt.ylabel("Records")
plt.tight_layout()
plt.show()
plt.savefig("../figures/missing_value_heatmap.png")
plt.close()
print("✅ Saved ../figures/missing_value_heatmap.png")

# --- 3. Outlier highlighting (boxplots) -------------------------------------
for col in numeric_cols:
    if col in df.columns:
        plt.figure(figsize=(6, 5))
        sns.boxplot(x=df[col])
        plt.title(f"Outlier Detection: {col}")
        plt.xlabel(col)
        plt.tight_layout()

        # Display inline
        plt.show()

        # Save figure
        plt.savefig(f"../figures/outlier_{col}.png")
        plt.close()
        print(f"✅ Saved ../figures/outlier_{col}.png")

# --- 4. Correlation heatmap --------------------------------------------------
plt.figure(figsize=(8, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Matrix: Vital Statistics")
plt.tight_layout()
plt.show()
plt.savefig("../figures/correlation_matrix.png")
plt.close()
print("✅ Saved ../figures/correlation_matrix.png")

print("\n🎯 Visualization complete — figures displayed and saved.")
```

```python
# ==========================================================
# 5. Summary Report & Notebook Conclusion
# ==========================================================
# Purpose:
#   - Summarize overall data quality findings
#   - Provide narrative insights for README or presentation use
#   - Save summary text to processed/summary_report.txt
# ==========================================================

import pandas as pd
import numpy as np
import os

# --- Load dataset ------------------------------------------------------------
df = pd.read_csv("../data/raw/synthetic_ehr_data.csv")

# --- Calculate high-level stats ---------------------------------------------
row_count, col_count = df.shape
missing_total = df.isna().sum().sum()
missing_ratio = missing_total / (row_count * col_count) * 100

# numeric distribution
numeric_cols = df.select_dtypes(include=[np.number]).columns
summary_stats = df[numeric_cols].describe().T

# --- Identify columns with most missing data ---------------------------------
missing_summary = df.isna().sum().sort_values(ascending=False)
top_missing = missing_summary.head(5)

# --- Build narrative summary -------------------------------------------------
narrative = f"""
=== Summary of Findings ===

1️⃣  Dataset Overview:
 • Records analyzed: {row_count:,}
 • Features included: {col_count}
 • Overall missing data: {missing_ratio:.2f}%

2️⃣  Data Completeness:
 • Columns with most missing values:
{top_missing.to_string()}

3️⃣  Value Ranges & Consistency:
Numeric feature summary (mean ± std):

{summary_stats[['mean', 'std']].round(2).head(6).to_string()}

4️⃣  Key Observations:
 • Most vital signs (blood pressure, cholesterol) are within realistic clinical ranges.
 • A small fraction of missing or extreme values suggests EHR export noise rather than systemic data issues.
 • The dataset is ready for downstream analytics—predictive modeling, readmission risk scoring, or clinical quality audits.

5️⃣  Next Steps:
 • Integrate additional data sources (lab results, diagnosis codes).
 • Validate patient identity linkage consistency.
 • Apply anomaly detection or clustering to discover outlier cohorts.
 • Automate daily data quality monitoring using Great Expectations or similar frameworks.

✅ Report generated successfully.
"""

# --- Print narrative --------------------------------------------------------
print(narrative)

# --- Save to file -----------------------------------------------------------
os.makedirs("../data/processed", exist_ok=True)
with open("../data/processed/summary_report.txt", "w") as f:
    f.write(narrative)

print("\n📝 Summary report saved → ../data/processed/summary_report.txt")
```
